# Pre-processing

```python
!pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
def WordEmbeddings(texts):

  model = SentenceTransformer('bert-large-nli-stsb-mean-tokens')

  sentence_embeddings = model.encode(texts)

  return sentence_embeddings 
```

# One Class Classification

**to run, it is advised to use GPU**

```python
def Classification(text):
  embeddings_test = WordEmbeddings([text])
  resp = clf.predict(embeddings_test)
  if resp[0] == 1:
    print('The text: ' + text + '\n belongs to the class of interest!') 
  if resp[0] == -1:
    print('The text: ' + text + '\n does not belong to the class of interest!') 
```

```python
for text in texts:
  Classification(text)
```
