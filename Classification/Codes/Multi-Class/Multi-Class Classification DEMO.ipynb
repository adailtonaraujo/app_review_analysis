First you must pré-processing the texts with the same method that the texts on the train step was pre-processed. We use the word-embedding RoBERTa.

# Pre-processing

```python
!pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
def WordEmbeddings(texts):

  model = SentenceTransformer('roberta-large-nli-stsb-mean-tokens')

  sentence_embeddings = model.encode(texts)

  return sentence_embeddings 
```

# Classification

**to run, it is advised to use GPU**

```python
def Classification(text):
  embeddings_test = WordEmbeddings([text])
  resp = clf.predict(embeddings_test)
  print('The text: ' + text + '\nbelongs to the '+ str(resp[0]) +' class' ) 
```

```python
for text in texts:
  Classification(text)
```
